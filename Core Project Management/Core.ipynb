```python
import pandas as pd

# Load the datasets
accounts = pd.read_csv('Accounts.csv')
customers = pd.read_csv('Customers.csv')
issues = pd.read_csv('Issues.csv')
users = pd.read_csv('Users.csv')
worklog = pd.read_csv('Worklogs.csv')

accounts_customers = pd.merge(accounts, customers, on='CustomerKey', how='outer')
accounts_customers.head()
```

```python
worklog_accounts_customers = pd.merge(worklog, accounts_customers, on='AccountKey', how='outer')
worklog_accounts_customers.head()
```

```python
issues_worklog_accounts_customers = pd.merge(worklog_accounts_customers, issues, left_on='IssueId', right_on='Issue id', how='outer')
issues_worklog_accounts_customers.head()
```

```python
final_data = pd.merge(issues_worklog_accounts_customers, users, on='UserKey', how='outer')
final_data.head()
```

```python
# Display the structure and preview of the final dataset
print(final_data.info())
print(final_data.head())

# Save the final dataset to a CSV file for further use
final_data.to_csv('Final_Joined_Dataset.csv', index=False)
```

```python
# Replacing values in the 'Beverage' column
replacements = {
    "Continuous" : "Ongoing",   
    "Done" : "Done",                          
    "In Progress" : "Ongoing", 
    "To Do" : "Upcoming",                    
    "Completed" : "Done",                    
    "In Review" : "Ongoing",                      
    "? Done" : "Done",
    "Backlog"   : "Upcoming",                    
    "Requests"  : "Upcoming",                   
    "? In Progress" : "Ongoing",                
    "Waiting to assign" : "Upcoming",        
    "Archived" : "Upcoming",               
    "Open"  : "Ongoing",                      
    "Events" : "Upcoming",                 
    "Testing"    : "Ongoing",                      
    "Cancelled"  : "Terminated",                     
    "?? Backlog" : "Upcoming",                   
    "Future Consideration" : "Upcoming",          
    "Immediate Action" : "Upcoming",          
    "Awaiting Response"  : "Ongoing",              
    "Review" : "Ongoing",                        
    "Proposed Priority Backlog" : "Upcoming",   
    "STCi" : "Ongoing",                       
    "Cancelled / Duplicate" : "Terminated",       
    "Project Summaries" : "Ongoing",           
    "? Cancelled" : "Terminated",                     
    "Rejected"  : "Terminated",                     
    "Upcoming External" : "Upcoming",            
    "Waiting" : "Ongoing"  
}

final_data["Status_y"] = final_data["Status_y"].replace(replacements)

final_data.head()
```

```python
final_data["Status_y"].head()
```

```python
# List of columns to convert to datetime
datetime_formats = {
    'startDate': '%d-%b-%y',
    'startTime': '%H:%M:%S',
    'createdAt': '%d/%m/%Y %H:%M',
    'updatedAt': '%d/%m/%Y %H:%M',
    'Created': '%d/%m/%Y %H:%M',
    'Updated': '%d/%m/%Y %H:%M',
    'Last Viewed': '%d/%m/%Y %H:%M',
    'Resolved': '%d/%m/%Y %H:%M',
    'Due date': '%d/%m/%Y %H:%M'
}

# Convert columns to datetime with specified formats
for column, fmt in datetime_formats.items():
    if column in final_data.columns:
        final_data[column] = pd.to_datetime(final_data[column], format=fmt, errors='coerce')

# Replace NaT with an empty string (if required)
final_data = final_data.fillna('')

# Display the first few rows to verify
print(final_data.head())
```

# DASHBOARD 1

### Step 1: Derive Number of Active Projects

```python
# Filter rows where Status indicates the project is ongoing
active_projects = final_data[final_data['Status_y'] == 'Ongoing']
number_of_active_projects = active_projects['ProjectKey'].nunique()  # Count unique projects
print(f"Number of Active Projects: {number_of_active_projects}")
```

```python
final_data['Status_y'].value_counts()
```

```python
final_data['Active_Projects'] = final_data['Status_y'].apply(lambda val: 1 if val == "Ongoing" else 0)
final_data['Active_Projects'].head()
```

```python
final_data['Active_Projects'].value_counts()
```

```python
final_data['resolved status'] = final_data['Resolved'].apply(lambda x: 1 if pd.notnull(x) else 0)

final_data['resolved status'].value_counts()
```

### Step 2: Derive Percentage of Late Projects

```python
# Filter rows where a project is marked as late
#late_projects = final_data[(final_data['Status_x'] == 'Late') | (issues['Resolved'] > issues['Due date'])]
late_projects = final_data[(final_data['Resolved'] > final_data['Due date'])]
number_of_late_projects = late_projects['ProjectKey'].nunique()  # Count unique late projects

# Total projects
total_projects = final_data['ProjectKey'].nunique()

# Calculate percentage of late projects
percentage_of_late_projects = (number_of_late_projects / total_projects) * 100 if total_projects > 0 else 0
print(f"Percentage of Late Projects: {percentage_of_late_projects:.2f}%")
```

```python
number_of_late_projects
```

```python
final_data['Late_Projects'] = (final_data['Resolved'] > final_data['Due date']).astype(int)
final_data['Late_Projects'].head()
```

```python
final_data['Late_Projects'].value_counts()
```

### Step 3: Derive Average Task Completion Time

```python
# Filter completed tasks (tasks with a non-null Resolved date)
completed_tasks = final_data[final_data['Resolved'].notnull()]

# Calculate task completion time for completed tasks
completed_tasks['TaskCompletionTime'] = (
    pd.to_datetime(completed_tasks['Resolved']) - pd.to_datetime(completed_tasks['Created'])
).dt.days

# Calculate average task completion time
average_task_completion_time = completed_tasks['TaskCompletionTime'].mean() if not completed_tasks.empty else 0
print(f"Average Task Completion Time: {average_task_completion_time:.2f} days")
```

```python
final_data['TaskCompletionTime'] = (final_data['Resolved'] - final_data['Created']).dt.days
final_data['TaskCompletionTime'].value_counts()
```

# DASHBOARD 2

```python
# Check the data types and convert columns to numeric where necessary
final_data['timeSpentSeconds'] = pd.to_numeric(final_data['timeSpentSeconds'], errors='coerce')
final_data['billableSeconds'] = pd.to_numeric(final_data['billableSeconds'], errors='coerce')

# Add calculated fields
# Check for NaN values after conversion and replace them with 0
final_data['timeSpentSeconds'] = final_data['timeSpentSeconds'].fillna(0)
final_data['billableSeconds'] = final_data['billableSeconds'].fillna(0)
```

```python
print(final_data[['timeSpentSeconds', 'billableSeconds']].dtypes)
```

```python
# Convert seconds to hours
final_data['TimeSpentHours'] = final_data['timeSpentSeconds'] / 3600
final_data['BillableHours'] = final_data['billableSeconds'] / 3600
final_data['NonBillableHours'] = final_data['TimeSpentHours'] - final_data['BillableHours']
```

```python
print(final_data.dtypes)
```

### Step 1: Calculate Average Utilization Rate (%)

```python
employee_utilization = final_data.groupby('UserKey').agg(
    TotalHours=('TimeSpentHours', 'sum'),
    BillableHours=('BillableHours', 'sum')
).reset_index()

employee_utilization['UtilizationRate(%)'] = (
    (employee_utilization['BillableHours'] / employee_utilization['TotalHours']) * 100
).fillna(0)

average_utilization_rate = employee_utilization['UtilizationRate(%)'].mean()
print(f"Average Utilization Rate (%): {average_utilization_rate:.2f}%")
```

```python
final_data['UtilizationRate(%)'] = ((final_data['BillableHours'] / final_data['TimeSpentHours']) * 100).fillna(0)
final_data['UtilizationRate(%)'].head()
```

### Step 2: Employee with the Highest Billable Hours

```python
top_employee = employee_utilization.loc[
    employee_utilization['BillableHours'].idxmax(), 'UserKey'
]
highest_billable_hours = employee_utilization['BillableHours'].max()
print(f"Employee with the Highest Billable Hours: UserKey {top_employee} ({highest_billable_hours:.2f} hours)")
```

### Step 3: Total Non-Billable Hours

```python
total_non_billable_hours = final_data['NonBillableHours'].sum()
print(f"Total Non-Billable Hours: {total_non_billable_hours:.2f} hours")
```

# DASHBOARD 3

```python
# Step 1: Total Hours Spent Per Category
time_per_category = final_data.groupby('CategoryName').agg(
    TotalHours=('TimeSpentHours', 'sum'),
    BillableHours=('BillableHours', 'sum')
).reset_index()
time_per_category
```

```python
# Step 2: Billable Hours Proportion Per Category (%)
time_per_category['BillableHours(%)'] = (
    (time_per_category['BillableHours'] / time_per_category['TotalHours']) * 100
).fillna(0)
time_per_category
```

```python
# Step 3: Category with the Most Time Spent
category_most_time = time_per_category.loc[
    time_per_category['TotalHours'].idxmax(), 'CategoryName'
]
most_time_hours = time_per_category['TotalHours'].max()
print(f"Category with the Most Time Spent: {category_most_time} ({most_time_hours:.2f} hours)")

# Print the category-wise details
print("\nTime Allocation by Category:")
print(time_per_category)
```
